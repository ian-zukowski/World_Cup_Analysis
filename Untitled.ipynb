# Join dataframes based on the winning teams, as the rows are misaligned currently

## Get a winning team column in import_matches2

```python
# Find winning team in matches2 - create a new column for it
def winner(row):
    home_name = row['Home Team Name']
    if row['Home Team Goals']>row['Away Team Goals']:
        return row['Home Team Name']
    if row['Away Team Goals']>row['Home Team Goals']:
        return row['Away Team Name']
    return f'Draw: {home_name}'
```

```python
import_matches2['Winning Team'] = import_matches2.apply(lambda row: winner(row), axis=1)
```

```python
import_matches3 = import_matches2.dropna()
```
