# Transform dataframes so similar rows actually match

```python
import_matches_a1 = import_matches_a.copy()
```

```python
def draw_replace_win(row):
    team = row['home_team']
    winner = row['winning_team']
    if row['outcome']=='D':
        return f'Draw: {team}'
    else:
        return winner
    
def draw_replace_loss(row):
    team = row['away_team']
    loser = row['losing_team']
    if row['outcome']=='D':
        return f'Draw: {team}'
    else:
        return loser
```

```python
import_matches_a1['winning_team'] = import_matches_a1.apply(lambda row: draw_replace_win(row), axis=1)
import_matches_a1['losing_team'] = import_matches_a1.apply(lambda row: draw_replace_loss(row), axis=1)
```

## Problem 1: Some rows have swapped home/away teams in 'import_matches_a' and 'import_matches_b'

```python
# Many rows have the home team and away teams swapped with each other
swapped_home_away_df = joined_df2.loc[(joined_df2['Home Team Name']==joined_df2['away_team'])|
                                      (joined_df2['Away Team Name']==joined_df2['home_team'])]
```

```python
# Those swapped rows are displayed -- show that the versions from import_matches7 should get switched
swapped_home_away_df[['Home Team Name','home_team', 'Away Team Name','away_team', 'Home Team Goals','home_score','Away Team Goals','away_score','winning_team']]
```

```python
# Create functions to swap the 4 home/away values imported from import_matches7
def swap_home_name(row):
    home_name = row['Home Team Name']
    return home_name

def swap_away_name(row):
    away_name = row['Away Team Name']
    return away_name

def swap_home_score(row):
    home_score = row['Home Team Goals']
    return home_score

def swap_away_score(row):
    away_score = row['Away Team Goals']
    return away_score
```

```python
# Apply those 4 functions on the columns from import_matches7
swapped_home_away_df['home_team'] = swapped_home_away_df.apply(lambda row: swap_home_name(row), axis=1)
swapped_home_away_df['away_team'] = swapped_home_away_df.apply(lambda row: swap_away_name(row), axis=1)
swapped_home_away_df['home_score'] = swapped_home_away_df.apply(lambda row: swap_home_score(row), axis=1)
swapped_home_away_df['away_score'] = swapped_home_away_df.apply(lambda row: swap_away_score(row), axis=1)
```

```python
# Verify the functions worked properly
swapped_home_away_df[['Home Team Name','home_team', 'Away Team Name','away_team', 'Home Team Goals','home_score','Away Team Goals','away_score','winning_team']].head()
```

```python
# Go back to the joined_df2 and actually swap those columns
joined_df2.loc[joined_df2['Home Team Name']==joined_df2['away_team']] = swapped_home_away_df
```
