```python
import_matches_b1 = import_matches_b.copy()
```

```python
# Find winning team in matches2 based on which team scored more goals
def winner(row):
    home_name = row['Home Team Name']
    if row['Home Team Goals']>row['Away Team Goals']:
        return row['Home Team Name']
    if row['Away Team Goals']>row['Home Team Goals']:
        return row['Away Team Name']
    return f'Draw: {home_name}'
```

```python
# Apply the function to create a new column referencing each match winner
import_matches_b1['Winning Team'] = import_matches_b1.apply(lambda row: winner(row), axis=1)
```
